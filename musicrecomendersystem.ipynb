```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\kalan\Downloads\music.csv")
df.head()
```

```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
df.head()
```

```python
l=[]
for i in df['User-Rating']:
    l.append(i[0:3])
```

```python
df['User-Rating']=l
```

```python
df['Singer/Artists']=df['Singer/Artists'].str.replace(' ','')
df['Album/Movie']=df['Album/Movie'].str.replace(' ','')
```

```python
df['Singer/Artists']=df['Singer/Artists'].str.replace(',',' ')
df['Album/Movie']=df['Album/Movie'].str.replace(',',' ')
```

```python
df.head()
```

```python
df['tags']=df['Singer/Artists']+' '+df['Genre']+' '+df['Album/Movie']+' '+df['User-Rating']
df['tags'][0]
```

```python
newdata=df[['Song-Name','tags']]
newdata.head()
```

```python
newdata['tags']=newdata['tags'].apply(lambda x:str(x).lower())
newdata
```

```python
newdata.dropna(inplace=True)
```

```python
newdata.isna().sum()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2000)
```

```python
vectors=cv.fit_transform(newdata['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
newdata.rename(columns={'Song-Name':'Title'},inplace=True)
```

```python
def recommend(music):
    music_index=newdata[newdata['Title']==music].index[0]
    distances=similarity[music_index]
    music_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in music_list:
        print(newdata.iloc[i[0]].Title)
```

```python
recommend('Coca Cola')
```
